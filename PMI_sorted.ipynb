```python
# make a function to do all of this. This function could be cleaner as a series of modular parts.
def FileHeadings(filename):
    import pandas as pd
    data = pd.read_csv(filename) #read in file
    data = data.drop('user',1) #drop the user column if necessary
    return data

def FileNoHeadings(filename):
    import pandas as pd
    data = pd.read_csv(filename, sep = '\t')
    return data

def MakePMI(matrixname):
    import pandas as pd
    from math import log
    data = FileNoHeadings(matrixname)
    # create data frame with column names on both axes
    crossmat = pd.DataFrame(index=data.columns, columns=data.columns)
    indexLen = len(data.index) #number of rows

    sing_probs = dict() #establish empty dictionary for singular probabilities
    # fill dictionary 
    for x in data:
        sing_probs[x] = sum(data[x])/float(indexLen)
    
    # Let's define some helper functions

    # calculates the PMI numerator
    def PMInum (col1, col2, row_len):
        return sum(col1*col2/float(row_len))

    # calculates the PMI denominator
    def PMIdenom (col1, col2):
        return col1*col2

    # calculates PMI quotient
    def PMI (num, denom):
        return log(num/denom, 2)
    
    # now to calculate PMI!!!
    column_names= list(crossmat.columns.values) # list of the column names

    # Loop through the columns
    for i in range(0,len(crossmat.columns)):
        # Loop through the columns for each column
        for j in range(0,len(crossmat.columns)):
            numerator = PMInum(data.ix[:,i], data.ix[:,j], indexLen) #calculate numerator of PMI
            denominator = PMIdenom(sing_probs[column_names[i]], sing_probs[column_names[j]]) #calculate denom. of PMI
            if numerator/denominator == 0: #avoid log(0)
                crossmat.ix[i,j] = -1
            else:
                #calculate PMI total
                crossmat.ix[i,j] = PMI(numerator, denominator)/(-log(numerator,2))
    return crossmat
```
